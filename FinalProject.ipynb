### __Dataset Overview__

Every high res image `x` has three low res versions of itself:
- depixelated by 1/2
- depixelated by 1/4
- depixelated by 1/6

The `ImageSuperResolution_Unsplash` contains two folders and one file. The `high_res` folder contains all the high res images, and there are `1254` of them. The `low_res` folder contains all the low res images. Since there are `1254` high res images and each has three low res versions, then there are `3762` images in the `low_res` folder.

The format is as follows:

The low res images of `4.jpg` (found in `high_res` folder) are `4_2.jpg`, `4_4.jpg`, and `4_6.jpg`. Where for any `i` $^{th}$ image `i_r.jpg` in the `low_res` folder, `1/r` is how much it is depixelated by.

To make things simpler and more efficient, the file `image_data.csv` contains the names of the low resolution images and their corresponding high resolution images.

Because we are loading images' data from pandas dataframe to PyTorch, we need to implement/customize/override `torch.utils.data.Dataset` (read more [here](https://pytorch.org/tutorials/beginner/data_loading_tutorial.html) and [here](https://stackoverflow.com/a/61393991)).

We also need to resize the data to avoid errors like [this](https://discuss.pytorch.org/t/runtimeerror-stack-expects-each-tensor-to-be-equal-size-but-got-3-224-224-at-entry-0-and-3-224-336-at-entry-3/87211) (which occured to me btw lol; i got this error and came across this solution).

```python
class ISR_Unsplash_Dataset(Dataset):
    def __init__(self, df, col, transform=None):
        '''
        Args:
            df: Dataframe.
            col: low_res or high_res column (str) 
        '''
        self.df = df
        self.col = col
        self.transform = transform
        
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, index):
        row = self.df.iloc[index]
        sample = io.imread(row[self.col])
        
        if self.transform:
            sample = self.transform(sample)
            
        return sample
    
class Rescale(object):
    def __init__(self, output_size):
        assert isinstance(output_size, (int, tuple))
        self.output_size = output_size
        
    def __call__(self, sample):
        image = sample
        
        height, width = image.shape[:2]
        if isinstance(self.output_size, int):
            if height > width:
                height_, width_ = self.output_size * height/width, self.output_size
            else:
                height_, width_ = self.output_size, self.output_size * width/height
        else:
            height_, width_ = self.output_size
            
        height_, width_ = int(height_), int(width_)
        img = transform.resize(image, (height_, width_))
        
        return img

class ToTensor(object):
    def __call__(self, sample):
        image = sample
        
        # swap color axis because
        # numpy image: H x W x C
        # torch image: C x H x W
        image = image.transpose((2, 0, 1))
        return torch.from_numpy(image)
```

```python
tsfrm = T.Compose([Rescale((800, 1200))])
NUM_TRAIN, NUM_VAL, NUM_TEST = 1000, 100, 154
batch_size = 4 # load 4 samples per batch


train_HIGHRES = ISR_Unsplash_Dataset(data, 
                                  'high_res',
                                  transform=tsfrm)
train_HIGHRES_Loader = DataLoader(train_HIGHRES, 
                                  batch_size=batch_size)
```

```python
for i_batch, sample_batched in enumerate(train_HIGHRES_Loader):
    fig, axs = plt.subplots(1, 4, figsize=(20,5))
    for i in range(batch_size):
        axs[i].imshow(sample_batched[i])
        axs[i].set_title('sample ' + str(i+1))
    if i_batch == 3:
        break
```
