### __Features details:__

- `senator` - Name of Senator (first last)
- `party` - 0: Republican, 1: Democrat, 2: Independent
- `sex` - 1: Male, 0: Female
- `trade_date` - Date of initial transaction
- `owner` - Owner of the trade (Spouse, Child, or Self=Senator)
- `ticker` - Stock Ticker
- `type` - Purchase or Sale
- `trade_amount_max` - Maximum amount the trade could have been
- `trade_amount_min` - Minimum amount the trade could have been
- `buy_one_year_change` - One year change in value of the stock purchased from the date of transaction
- `one_mo_vtsax_change` - One month change in value of the VTSAX from the date of transaction
- `one_year_vtsax_change` - One year change in value of the VTSAX from the date of transaction
- `buy_compare_one_year` - Change in stock value minus change in VTSAX after one year from date of transaction
- `one_month_sell` - One month change in value of the stock sold from the date of transaction
- `one_month_sell_compare` - Change in stock value minus change in VTSAX after one month from date of transaction
- `weight` - Weighted value of the average between the minimum and maximum amount the trade could have been
